```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from functools import reduce
```

```python
df_Bitcoin = pd.read_csv('Bitcoin.csv')
df_Bitcoin.head()
```

```python
#def to calculate if the price increased or decreased 
df_Bitcoin['Change'] = df_Bitcoin["Close"] - df_Bitcoin["Open"]
```

```python
def change(x):
    if x > 0: 
        return 1 
    else: 
        return 0
        
df_Bitcoin["New_Change"] = df_Bitcoin["Change"].apply(change)
df_Bitcoin
```

```python
df_Bitcoin["New_Change"].value_counts()
#1024/1827 = 56% baseline accuracy 
```

```python
df_Bitcoin.rename(columns = {'Open':'Bitcoin_Open', 'High':'Bitcoin_High',
                              'Low':'Bitcoin_Low', 'Close':'Bitcoin_Close','Adj Close':'Bitcoin_Adj_Close','Volume':'Bitcoin_Volume','Change':'Bitcoin_Daily_Change','New_Change':'Price_Increase' }, inplace = True)
df_Bitcoin.head()
```

```python
df_NASDAQ = pd.read_csv('NASDAQ (1).csv')
df_NASDAQ.rename(columns = {'Open':'NASDAQ_Open', 'High':'NASDAQ_High',
                              'Low':'NASDAQ_Low', 'Close':'NASDAQ_Close','Adj Close':'NASDAQ_Adj_Close','Volume':'NASDAQ_Volume' }, inplace = True)
df_NASDAQ.head()
```

```python
df_Volatility = pd.read_csv('Volatility.csv')
df_Volatility.rename(columns = {'Open':'Volatility_Open', 'High':'Volatility_High',
                              'Low':'Volatility_Low', 'Close':'Volatility_Close','Adj Close':'Volatility_Adj_Close','Volume':'Volatility_Volume' }, inplace = True)
df_Volatility.head()
```

```python
df_Volatility.drop(['Volatility_Volume'], 1, inplace=True)
```

```python
df_Gold = pd.read_csv('Gold.csv')
df_Gold.rename(columns = {'Open':'Gold_Open', 'High':'Gold_High',
                              'Low':'Gold_Low', 'Close':'Gold_Close','Adj Close':'Gold_Adj_Close','Volume':'Gold_Volume' }, inplace = True)
df_Gold.head()
```

```python
df_10year = pd.read_csv('10year.csv')
df_10year.rename(columns = {'Open':'10year_Open', 'High':'10year_High',
                              'Low':'10year_Low', 'Close':'10year_Close','Adj Close':'10year_Adj_Close','Volume':'10year_Volume' }, inplace = True)
df_10year.head()
```

```python
df_GoogleSearch = pd.read_csv('GoogleSearch.csv')
df_GoogleSearch.rename(columns = {'Label': 'Date', 'Value': 'Google_Search_Volume'}, inplace = True)
df_GoogleSearch.head()
```

```python
dfs = [df_Bitcoin,df_Volatility,df_NASDAQ,df_Gold,df_10year,df_GoogleSearch]
```

```python
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), dfs)

# if you want to fill the values that don't exist in the lines of merged dataframe simply fill with required strings as

df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['Date'],
                                            how='outer'), dfs).fillna(0) # filled missing values with 0
```

```python
df = df_merged 
```

```python
df.to_csv('Bitcoin_Final_Project.csv')
```
